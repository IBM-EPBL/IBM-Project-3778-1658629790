## 9. CLUSTERING ALGORITHM

```python
x = data.iloc[:, [3, 4]].values 
```

```python
from sklearn.cluster import KMeans  
wcss_list= []
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 40)  
    kmeans.fit(x)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss_list)  
plt.title('The Elobw Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()  
```

## 10. ADD THE CLUSTER DATA WITH PRIMARY DATA

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x) 
m=pd.Series(y_predict)
data["Cluster"]=m
data
```

## 12. SPLIT THE DATA INTO TRAINING AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
y = data["Age"]
x = data.drop(columns = ["Age"])
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

## 13. BUILD THE MODEL

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

## 14. TRAIN THE MODEL

```python
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix
```

```python
pred = model.predict(x_train)
pred
```

```python
print("Accuracy score :",accuracy_score(y_train,pred))
print("Precision score :",precision_score(y_train,pred,average="macro"))
```

```python
print(confusion_matrix(y_train,pred))
```

## 15. TEST THE MODEL

```python
pred = model.predict(x_test)
pred
```

```python
print("Accuracy score :",accuracy_score(y_test,pred))
print("Precision score :",precision_score(y_test,pred,average="macro"))
```

```python
print(confusion_matrix(y_test,pred))
```
